```python

class College:
    def __init__(self, name):
        self.name = name


class Student(College):
    def __init__(self, name, rollno, section):
        self.name = name
        self.rollno = rollno
        self.section = section


class AcademicMark(Student):
    def __init__(self, name, rollno, section, marks):
        Student.__init__(self, name, rollno, section)
        self.marks = marks


class SportsMark(Student):
    def __init__(self, name, rollno, section, sports_marks):
        Student.__init__(self, name, rollno, section)
        self.sports_marks = sports_marks

class StudentReport(AcademicMark, SportsMark):
    def __init__(self, name, rollno, section, marks, sports_marks):
        AcademicMark.__init__(self, name, rollno, section, marks)
        SportsMark.__init__(self, name, rollno, section, sports_marks)

    def display_report(self):
        print(f"Student Name: {self.name}")
        print(f"Roll No: {self.rollno}")
        print(f"Section: {self.section}")
        print(f"Academic Marks: {self.marks}")
        print(f"Sports Marks: {self.sports_marks}")


student_report = StudentReport("Jayarakash", "23cse586", "I", 89, 97
                               )
student_report.display_report()
```

```python
class Student:
    def get_student_details(self):
        self.name = input("Enter Name: ")
        self.roll = input("Enter Roll No: ")
        self.section = input("Enter Section: ")

class Academic(Student):
    def get_academic_marks(self):
        print("\nEnter Academic Marks")
        self.os = int(input("OS: "))
        self.java = int(input("Java: "))
        self.python = int(input("Python: "))
        self.cpp = int(input("C++: "))

    def grade(self, mark):
        mark = int(mark)
        if mark >= 90:
            return 'A'
        elif mark >= 75:
            return 'B'
        elif mark >= 60:
            return 'C'
        elif mark >= 50:
            return 'D'
        else:
            return 'E'

class Sports(Academic):
    def get_sports_marks(self):
        self.sports = int(input("\nEnter Sports Marks: "))

class Result(Sports):
    def display_result(self):
        total = self.os + self.java + self.python + self.cpp + self.sports
        average = total / 5

        print("\nFINAL RESULT")
        print("Name:", self.name)
        print("Roll No:", self.roll)
        print("Section:", self.section)

        print("\nSubjects Marks & Grades")
        print("OS:", self.os, "Grade:", self.grade(self.os))
        print("Java:", self.java, "Grade:", self.grade(self.java))
        print("Python:", self.python, "Grade:", self.grade(self.python))
        print("C++:", self.cpp, "Grade:", self.grade(self.cpp))
        print("Sports:", self.sports, "Grade:", self.grade(self.sports))

        print("\nTotal Marks:", total)
        print("Average:", average)
        print("Final Grade:", self.grade(average))
r = Result()
r.get_student_details()
r.get_academic_marks()
r.get_sports_marks()
r.display_result()
```

```python
class Abs:
    class Vehicle:
        def start(self):
            print("Vehicle started")

        def stop(self):
            print("Vehicle stopped")

    class Car(Vehicle):
        def start(self):
            print("Car started")

        def stop(self):
            print("Car stopped")

    class Bike(Vehicle):
        def start(self):
            print("Bike started")

        def stop(self):
            print("Bike stopped")

car = Abs.Car()
bike = Abs.Bike()
p=[car,bike]
for i in p:
    i.start()
    i.stop()
```
